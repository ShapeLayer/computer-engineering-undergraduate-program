# MLP로 XOR 학습 모델 생성

## 모델 정의

```python
import tensorflow as tf
```

```python
model = tf.keras.models.Sequential()
```

XOR: 두개의 입력 $\Rightarrow$ 한개의 출력

**(input) Dense Layer**
- $2$ size vector $\Rightarrow$ $2$ size vector

```python
model.add(tf.keras.layers.Dense(units=2, input_shape=(2,), activation='sigmoid'))
```

**(output) Dense Layer**
- $2$ size vector $\Rightarrow$ $1$ size vector

```python
model.add(tf.keras.layers.Dense(units=1, activation='sigmoid'))
```

```python
model.compile(loss='mean_squared_error', optimizer=tf.keras.optimizers.SGD(learning_rate=0.3))
```

```python
model.summary()
```

## 모델 사용

```python
import numpy as np
```

```python
X = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])
y = np.array([[0], [1], [1], [0]])
```

```python
model.fit(X, y, batch_size=1, epochs=10000)
```

```python
print(model.predict(np.array([[0, 1]])))
print(model.predict(X))
```
