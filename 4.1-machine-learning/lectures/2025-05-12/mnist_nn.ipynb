```python
import torch
import torchvision.datasets as dsets
import torchvision.transforms as transforms
import random
```

```python
print(
    torch.backends.mps.is_built(),
    torch.backends.mps.is_available()
)
```

```python
device = torch.device("mps")
```

```python
random.seed(0)
torch.manual_seed(0)
```

```python
learning_rate = .001
training_epochs = 15
batch_size = 100
```

```python
mnist_train = dsets.MNIST(root='dset/', train=True, transform=transforms.ToTensor(), download=True)
mnist_test = dsets.MNIST(root='dset/', train=False, transform=transforms.ToTensor(), download=True)
```

```python
# DataLoader가 필요한 이유: 데이터 로드할 때 값의 순서를 섞어서 학습시킬 수 있음
data_loader = torch.utils.data.DataLoader(dataset=mnist_train, batch_size=batch_size, shuffle=True, drop_last=True)
```

## Make nn Layers

```python
linear1 = torch.nn.Linear(784, 256, bias=True)
# Bias: 편향.. 편향이 없으면 원점을 지나는 직선을 그림
linear2 = torch.nn.Linear(256, 256, bias=True)
linear3 = torch.nn.Linear(256, 10, bias=True)
relu = torch.nn.ReLU()
```

```python
model = torch.nn.Sequential(linear1, relu, linear2, relu, linear3).to(device)
```

```python
criterion = torch.nn.CrossEntropyLoss().to(device)
optimizer = torch.optim.Adam(model.parameters(), lr=learning_rate)
```

```python
total_batch = len(data_loader)
for epoch in range(training_epochs):
    avg_cost = 0

    for X, y in data_loader:
        X = X.view(-1, 28 * 28).to(device)
        y = y.to(device)

        optimizer.zero_grad()
        hypothesis = model(X)
        # print(hypothesis.shape)  # == [100, 10]
        cost = criterion(hypothesis, y)
        cost.backward()
        optimizer.step()

        avg_cost += cost / total_batch
    
    print(f'Epoch: {epoch + 1}, cost = {avg_cost:.9f}')

print('Learning Finished!')
```

```python
with torch.no_grad():
    X_test = mnist_test.data.view(-1, 28 * 28).float().to(device)
    y_test = mnist_test.targets.to(device)

    prediction = model(X_test)
    correct_prediction = torch.argmax(prediction, 1) == y_test
    accuracy = correct_prediction.float().mean()

    r = random.randint(0, len(mnist_test) - 1)
    # print(mnist_test.data[r] == mnist_test.data[r:r+1])
    # print(mnist_test.data[r].shape, mnist_test.data[r:r+1].shape)
    # [28, 28], [1, 28, 28]
    X_each = mnist_test.data[r].view(-1, 28 * 28).float().to(device)
    print(f'X_each shape: {X_each.shape}')
    y_each = mnist_test.targets[r].to(device)
    print(f'Label: {y_each.item()}')
    print(f'Prediction: {torch.argmax(prediction[r]).item()}')
    prediction_each = model(X_each)
    print(f'Accuracy: {accuracy.item():.4f}')
```

```python
model = torch.nn.Sequential(
    torch.nn.Linear(784, 1024, bias=True),
    torch.nn.ReLU(),
    torch.nn.Linear(1024, 1024, bias=True),
    torch.nn.SELU(),
    torch.nn.Linear(1024, 512, bias=True),
    torch.nn.ReLU(),
    torch.nn.Linear(512, 256, bias=True),
    torch.nn.SELU(),
    torch.nn.Linear(256, 10, bias=True)
).to(device)
```

```python
criterion = torch.nn.CrossEntropyLoss().to(device)
optimizer = torch.optim.Adam(model.parameters(), lr=learning_rate)
```

```python
total_batch = len(data_loader)
for epoch in range(training_epochs):
    avg_cost = 0

    for X, y in data_loader:
        X = X.view(-1, 28 * 28).to(device)
        y = y.to(device)

        optimizer.zero_grad()
        hypothesis = model(X)
        cost = criterion(hypothesis, y)
        cost.backward()
        optimizer.step()

        avg_cost += cost / total_batch

    print(f'Epoch: {epoch + 1}, cost = {avg_cost:.9f}')

print('Learning Finished!')
```

```python
with torch.no_grad():
    X_test = mnist_test.data.view(-1, 28 * 28).float().to(device)
    y_test = mnist_test.targets.to(device)

    prediction = model(X_test)
    correct_prediction = torch.argmax(prediction, 1) == y_test
    accuracy = correct_prediction.float().mean()
    print(f'Accuracy: {accuracy.item():.4f}')
```
