<h4>
    11. 前処理 전처리
</h4>

「train_cafe.csv」と「prob_cafe.csv」を読み込み，質的データに対しては次のラベルエンコーディングを行います．  
「train_cafe.csv」와 「prob_cafe.csv」를 읽어들여, 질적 데이터에 대해서는 다음의 라벨 인코딩을 수행합니다.

「月→0」, 「火→1」, 「水→2」, 「木→3」, 「金→4」  
「定食→0」, 「丼・カレー→1」, 「麺類→2」, 「前日不明→3」

その後，空席率の欠損値は中央値で埋めます．  
그 후, 공석률의 결측치는 중앙값으로 채웁니다.

欠損値を埋めた中央値の値を回答してください．  
결측치를 채운 중앙값의 값을 답변하세요.

```python
import pandas as pd

train = pd.read_csv('./assets/train_cafe.csv')
prob = pd.read_csv('./assets/prob_cafe.csv')

day = {
    each: i
    for i, each in enumerate('月火水木金')
}

food = {
    each : i
    for i, each in enumerate(['定食', '丼・カレー', '麺類', '前日不明'])
}

# label encoding
train['曜日'] = train['曜日'].map(day)
prob['曜日'] = prob['曜日'].map(day)
train['前日メニュー'] = train['前日メニュー'].map(food)
prob['前日メニュー'] = prob['前日メニュー'].map(food)
train['おすすめ'] = train['おすすめ'].map(food)
prob['おすすめ'] = prob['おすすめ'].map(food)
train['メニュー'] = train['メニュー'].map(food)

# fill missing values
train['空席率'] = train['空席率'].fillna(train['空席率'].median())
```

<h4>
    12.ハイパーパラメータ探索 <br />
    &nbsp; &nbsp; 하이퍼파라미터 탐색
</h4>

次をプログラミングしてください．
다음을 프로그래밍하세요.

モデルはランダムフォレストを使います．
모델은 랜덤 포레스트를 사용합니다.

ランダムフォレストのハイパーパラメータは n_estimators=401 を固定し，max_depth と min_samples_leaf をグリッドサーチの対象とします．それ以外のハイパーパラメータはデフォルトを使用します．  
랜덤 포레스트의 하이퍼파라미터는 n_estimators=401로 고정하고, max_depth와 min_samples_leaf를 그리드 서치의 대상으로 합니다. 그 외의 하이퍼파라미터는 기본값을 사용합니다.

max_depth は [2, 4, 8, 16, 32] を，min_samples_leaf は [1, 5, 10] をグリッドサーチし，正解率が最も高いモデルを探します．なお，交差検証は5回とします．
max_depth는 [2, 4, 8, 16, 32]를, min_samples_leaf는 [1, 5, 10]를 그리드 서치하여, 정답률이 가장 높은 모델을 찾습니다. 단, 교차 검증은 5회로 합니다.

このとき，最良モデルの検証スコアが入る階級を選んでください．
이 때, 최상의 모델의 검증 스코어가 들어가는 계급을 선택하세요.

なお，ランダムフォレストは乱数を使うので実行する度に結果が変わります．どの階級に入るか迷った場合は何度かプログラムを実行し，最も確信が高い階級を選択しましょう．  
또한, 랜덤 포레스트는 난수를 사용하므로 실행할 때마다 결과가 바뀝니다. 어느 계급에 들어갈지 망설여질 경우에는 몇 번이고 프로그램을 실행하여, 가장 확신이 높은 계급을 선택합시다.

- 0.65 未満
- 0.65 以上 0.70 未満
- 0.70 以上 0.75 未満
- 0.75 以上 0.80 未満
- 0.80 以上 0.85 未満
- 0.85 以上 0.90 未満
- 0.90 以上 0.95 未満
- 0.95 以上 1.0 以下

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV

params = {
    'max_depth': [2, 4, 8, 16, 32],
    'min_samples_leaf': [1, 5, 10]
}
X_train = train[['曜日', '前日メニュー', '空席率', '残り時間', 'おすすめ', ]]
y_train = train['メニュー']

for each in range(5):
    rf = RandomForestClassifier(n_estimators=401, random_state=42)
    grid_search = GridSearchCV(rf, params, cv=5, scoring='accuracy')
    grid_search.fit(X_train, y_train)

    print(f"best_score: {each}: {grid_search.best_score_}")
```

#### 回答 :
- $0.86$
- 0.85 以上 0.90 未満

Ⅰ. 사건·사례의 개요 <!--事件・事例の概要-->

지난 2월 2일, 한국에서 AI로 생성한 영상을 유튜브에 게시한 30대 남성이 현지의 전기통신기본법 등을 위반한 혐의로 구속되었다. 이 남성은 '순찰 24시'라는 유튜브 채널을 운영하며 경찰 보디캠으로 찍은 듯한 가짜 영상을 올려왔다.

Ⅱ. AI의 개요 <!-- AIの概要 -->

영상에서는 경찰관이 담배 피우는 학생들을 계도하거나, 여성 탈의실에 여장남자가 등장, 경찰관이 폭력을 행사하는 듯한 표현 등 자극적인 소재가 담겨있었다. 더 나아가 사회의 공분을 산 사건의 뉴스 장면을 AI로 흉내내어 실제 현장에 출동한 경찰관의 보디캠처럼 보이도록 제작하기까지 하여 의도적으로 사람들이 현실과 혼동하도록 만들었다.

이 영상들은 인스타그램, 틱톡, 페이스북 등 소셜 미디어에 숏폼 형태로 유포되어 누적 3천만이 넘는 조회수를 기록했다. 시민들은 이 영상들을 실제라고 인지한 채 댓글을 달거나, 과잉 체포를 이유로 실제로 경찰에 항의하는 사례도 발생했다. 경찰은 이러한 영상들이 공권력에 대한 국민적 신뢰를 훼손, 다른 허위 영상물을 제작, 유포해 더 큰 추가 피해가 발생할 수도 있다고 판단해 이 남성을 입건해 구속했다. 수사 과정에서는 이 남성이 금융 사기, 음란물 제작·유포 등으로도 수익을 올린 사실이 추가로 확인되었다. 해외의 유료구독형 SNS 채널을 운영하면서 AI 기술로 생성한 딥페이크 음란물을 제작하고 판매하는 방식이다.

Ⅲ. 해당하는 AI 윤리 과제 <!-- 該当するAI倫理の課題 -->

이 사례는 AI 윤리에서 '악의적인 활용' <!-- ⑥悪意ある活用 --> 에 해당한다. 직접적으로는 수업에서 악의적인 활용의 예시로서 소개된 '2024년 홍콩에서의 딥페이크 피싱 사건'과 같이 딥페이크 기술을 사용해 사람들의 인식을 의도적으로 비틀어 실현한 범죄라는 점에서 유사하다. 홍콩 건과는 다르게 직접적인 피해자는 발생하지는 않았으나, AI 생성물을 개인적인 목적에서 악의적으로 생성, 유포하여 금전적 이득을 취하고, 사회적으로 혼란과 피해를 야기했다.

Ⅳ. 재발 방지책 <!-- 再発防止策 -->

경찰의 대응에서 확인할 수 있듯 대개의 국가에서는 이미 AI 등장 이전의 사회 규범과 법률 체계로도 악의적인 AI사용을 처벌하고 대응할 수 있다. 규칙 면에서 이러한 사건사고는 잘 대응되고 있기 때문에, 개별 사용자에게 책임있는 AI 사용을 요구하는 것 이상으로 추가적인 대응은 한계가 명확할 것으로 보인다. 경찰에서는 허위정보 관련 범죄에 대해 집중단속을 전개하던 중에 이 남성의 범죄 사실을 확인하고 검거, 구속했다고 밝혔다. 이로 미루어보아 가용한 추가 대책은 법률과 규칙의 허점을 찾아 보완입법을 추진하고, 이 사례와 같이 가용 자원을 더 투입해 적극적으로 단속하는 것 까지일 것으로 보인다.

기술적인 면에서도 할 수 있는 일은 많지 않다. AI로 생성된 영상물임을 자동으로 탐지하는 기술이 널리 상용화되지 못했고, 시간이 지남에 따라 계속해서 AI들의 성능이 발전하고 있으므로 AI 생성물 여부를 파악하는 것이 쉽지 않다. AI 생성물에 워터마크를 삽입하는 것이 방법이 될 수 있겠으나, AI 모델은 공개되어 있어 비용만 지불하면 누구나 AI를 재현할 수 있고, 생산성을 향상한다는 AI 사용 목적에도 부합하지 않는다. 따라서 지속적인 자원 투입으로 규칙과 법률의 보완, 그리고 단속 강화가 이 사례와 같은 악의적인 AI 활용에 대한 현실적인 대응책이 될 것으로 판단된다.

Ⅰ. 事件・事例の概要

去る2月2日、韓国でAI生成動画をYouTubeに投稿していた30代の男が、現地の電気通信基本法違反などの疑いで拘束された。この男は「パトロール24時（순찰 24시）」というYouTubeチャンネルを運営し、警察のボディカムで撮影されたかのような偽の動画を投稿し続けていた。

Ⅱ. AIの概要

動画には、警察官が喫煙する学生を指導する場面や、女性脱衣所に女装した男が登場する場面、警察官が暴力を振るうような描写など、刺激的な素材が含まれていた。さらに、社会的な公憤を買った事件のニュース映像をAIで模倣し、実際の現場に出動した警察官のボディカム映像であるかのように制作することで、意図的に現実と混同させるよう仕向けていた。

これらの動画は、Instagram、TikTok、FacebookなどのSNSでショート動画形式で拡散され、累計3,000万回を超える再生回数を記録した。市民はこれらの映像を事実として認識してコメントを残したり、過剰な逮捕を理由に実際に警察へ抗議したりする事例も発生した。警察は、これらの動画が公権力に対する国民の信頼を毀損し、さらなる虚偽動画の制作・流布による二次被害を招く恐れがあると判断し、この男を立件・拘束した。捜査の過程では、この男が金融詐欺やわいせつ物の制作・流布でも利益を得ていた事実が追加で確認された。海外の有料購読型SNSチャンネルを運営しながら、AI技術で生成したディープフェイクわいせつ物を制作・販売する手法である。

Ⅲ. 該当するAI倫理の課題

本事例は、AI倫理における「悪意ある活用（⑥悪意ある活用）」に該当する。直接的には、講義で悪意ある活用の例として紹介された「2024年の香港におけるディープフェイク・フィッシング事件」と同様に、ディープフェイク技術を用いて人々の認識を意図的に歪め、犯罪を実現したという点で類似している。香港の事例とは異なり、直接的な被害者は発生していないものの、個人的な目的でAI生成物を悪意を持って生成・流布し、金銭的利益を得ただけでなく、社会的な混乱と被害を引き起こした。

Ⅳ. 再発防止策

警察の対応からも分かる通り、大半の国ではすでにAI登場以前の社会規範や法律体系によって、悪意あるAIの使用を処罰し、対応することが可能である。規則の面ではこうした事件・事故への対応は進んでおり、個々のユーザーに責任あるAI利用を求めること以上の追加対応には限界があると考えられる。警察は、虚偽情報関連犯罪の集中取り締まりを展開する中で、この男の犯罪事実を確認し検挙・拘束したと発表した。このことから、可能な追加対策としては、法律や規則の不備を見つけて補完立法を推進し、本事例のように利用可能なリソースをさらに投入して積極的に取り締まることまでが限界であると思われる。

技術的な面でも、できることは多くない。AI生成動画を自動的に検知する技術は広く実用化されておらず、時間の経過とともにAIの性能が向上し続けているため、AI生成物かどうかを判別することは容易ではない。AI生成物にウォーターマークを挿入することも一つの方法ではあるが、AIモデルは公開されており、コストさえ払えば誰でもAIを再現できる。また、生産性を向上させるというAIの使用目的にもそぐわない。したがって、継続的なリソース投入による規則と法律の補完、そして取り締まりの強化が、本事例のような悪意あるAI活用に対する現実的な対応策になると判断される。
